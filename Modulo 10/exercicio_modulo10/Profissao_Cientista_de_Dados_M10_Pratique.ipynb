```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('ecommerce_preparados.csv')

print(df.head(10).to_string())
print(df.tail(10).to_string())
```

**Para facilitar a visualização, optei por remover as colunas de Reviews, que não vou utilizar na análise.**

```python
df.drop(['Review1', 'Review2', 'Review3'], axis=1, inplace=True)
print(df.head(10).to_string())
print(df.tail(10).to_string())
```

```python
df.describe()
```

**O primeiro gráfico é sobre a coluna de Notas. Fiz primeiro um histograma para verificar possíveis picos na distribuição.**

```python
plt.figure(figsize=(8, 4))
plt.hist(df['Nota'], bins=30, color='red', alpha=0.8, edgecolor='pink')
plt.xticks(ticks=range(2, int(df['Nota'].max() + 1), 1))
plt.yticks(ticks=range(0, 250, 20))
plt.title('Histograma - Distribuição de Notas')
plt.xlabel('Notas')
plt.ylabel('Frequência')
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
plt.show()
```

**Em seguida, criei um gráfico de densidade a fim de melhor visualizar a distribuição das notas.**

```python
plt.figure(figsize=(8, 4))
sns.kdeplot(df['Nota'], fill=True, color='red')
plt.title('Densidade de Notas')
plt.xlabel('Nota')
plt.ylabel('Densidade')
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
plt.show()
```

**Começando a analisar a coluna de gênero, contei quantos valores únicos ela tinha e pedi para listar quais eram eles.**

```python
print(f"Número de valores únicos em 'Gênero': {df['Gênero'].nunique()}")
print(f"Lista de valores em 'Gênero': {df['Gênero'].unique()}")
```

**Notando que várias dos valores eram sobrepostos, considerei ser melhor renomear alguns deles para diminuir o número de categorias para o gráfico.**

```python
df['Gênero'] = df['Gênero'].replace('Unissex', 'Sem gênero')
df['Gênero'] = df['Gênero'].replace('Meninas', 'Feminino Infantil')
df['Gênero'] = df['Gênero'].replace('Meninos', 'Masculino Infantil')
df['Gênero'] = df['Gênero'].replace('Unissex', 'Sem gênero')
df['Gênero'] = df['Gênero'].replace('short menina verao look mulher', 'Feminino')
df['Gênero'] = df['Gênero'].replace('Mulher', 'Feminino')
df['Gênero'] = df['Gênero'].replace('roupa para gordinha pluss P ao 52', 'Feminino')
df['Gênero'] = df['Gênero'].replace('bermuda feminina brilho Blogueira', 'Feminino')
df['Gênero'] = df['Gênero'].replace('menino', 'Masculino Infantil')
```

**Agora, escolhi o gráfico de pizza para visualizar a proporção dessas categorias.**

```python
plt.figure(figsize=(8, 4))
x = df['Gênero'].value_counts().index
y = df['Gênero'].value_counts().values
plt.pie(y, labels=x, autopct='%1.1f%%')
plt.title('Distribuição de Gênero')
plt.show()
```

**E um gráfico de barras para verificar a quantidade de cada uma das categorias.**

```python
plt.figure(figsize=(8, 4))
x = df['Gênero'].value_counts().index
y = df['Gênero'].value_counts().values
plt.bar(x, y, color='#60aa65')
plt.title('Distribuição de Gênero')
plt.xlabel('Gênero')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.show()
```

**Aqui fiz um tratamento semelhante ao da coluna de Gênero para a coluna de Temporada, reduzindo valores redundantes.**

```python
print(f"Número de valores únicos em 'Temporada': {df['Temporada'].nunique()}")
print(f"Lista de valores em 'Temporada': {df['Temporada'].unique()}")
```

```python
df['Temporada'] = df['Temporada'].replace('primavera-verão outono-inverno', 'não definido')
df['Temporada'] = df['Temporada'].replace('primavera/verão/outono/inverno', 'não definido')
df['Temporada'] = df['Temporada'].replace('2021', 'não definido')
df['Temporada'] = df['Temporada'].replace('primavera-verão - outono-inverno', 'não definido')
df['Temporada'] = df['Temporada'].replace('primavera/verão outono/inverno', 'não definido')
df['Temporada'] = df['Temporada'].replace('outono-inverno', 'outono/inverno')
df['Temporada'] = df['Temporada'].replace('primavera-verão', 'primavera/verão')
```

```python
plt.figure(figsize=(8, 4))
x = df['Temporada'].value_counts().index
y = df['Temporada'].value_counts().values
plt.pie(y, labels=x, autopct='%1.1f%%')
plt.title('Distribuição de Temporada')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
x = df['Temporada'].value_counts().index
y = df['Temporada'].value_counts().values
plt.bar(x, y, color='blue')
plt.title('Distribuição de Temporada')
plt.xlabel('Temporada')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.show()
```

**Para a coluna de Marcas, primeiro verifiquei a quantidade de valores únicos e os listei.**

```python
print(f"Número de valores únicos em 'Marca': {df['Marca'].nunique()}")
print(f"Lista de valores em 'Marca': {df['Marca'].unique()}")
```

**Entretanto, o número foi muito elevado. Assim, solicitei a uma IA para criar um script para tentar fazer uma redução de valores redundantes como feito anteriormente.**

```python
df['Marca'] = df['Marca'].replace(['roupa', 'vestuário', 'roupas', 'conjunto', 'kit'], 'Roupa', regex=True)
df['Marca'] = df['Marca'].replace(['infantil', 'kids', 'baby'], 'Infantil', regex=True)
df['Marca'] = df['Marca'].replace(['fitness', 'academia', 'esportes', 'sport', 'sports'], 'Esporte', regex=True)
df['Marca'].replace({'multimarcas': 'Várias Marcas', 'variados': 'Várias Marcas'}, inplace=True)

# Corrigindo erros de digitação (ajuste conforme necessário)
df['Marca'] = df['Marca'].replace('puma', 'Puma', regex=True)

# Simplificando descrições
df['Marca'] = df['Marca'].replace('camiseta dry fit masculinas', 'Camiseta', regex=True)
df['Marca'] = df['Marca'].replace('short feminino', 'Short', regex=True)

# Outros exemplos (adapte conforme sua necessidade):
df['Marca'] = df['Marca'].replace('jeans', 'Jeans', regex=True)
df['Marca'] = df['Marca'].replace('lingerie', 'Lingerie', regex=True)

# Substituindo termos muito genéricos por "Outro"
df['Marca'] = df['Marca'].replace(['genérica', 'própria', 'nacional'], 'Outro', regex=True)

# Criando uma função para substituições mais complexas
def custom_replace(text):
    if 'atacado' in text:
        return 'Atacado'
    elif 'baby' in text:
        return 'Infantil'
    else:
        return text

df['Marca'] = df['Marca'].apply(custom_replace)
```

```python
print(f"Número de valores únicos em 'Marca': {df['Marca'].nunique()}")
```

**Como ainda assim o conjunto continuou grande, considerei melhor analisar somente as 10 mais vendidas.**

```python
contagem_marcas = df['Marca'].value_counts()
top_10_marcas = contagem_marcas.head(10)
print(top_10_marcas)
```

**Por fim, um gráfico de dispersão relacionando o preço com o tamanho dos descontos oferecidos.**

```python
sns.jointplot(x='Preço', y='Desconto', data=df, kind='scatter', color='#6a5acd')
plt.xlabel('Preço')
plt.ylabel('Desconto')
plt.gcf().suptitle("Relação entre Preço e Desconto")
plt.show()
```
