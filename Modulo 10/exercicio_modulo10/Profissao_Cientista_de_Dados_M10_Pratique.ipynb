```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('ecommerce_preparados.csv')

print(df.head(10).to_string())
print(df.tail(10).to_string())
```

**Para facilitar a visualiza√ß√£o, optei por remover as colunas de Reviews, que n√£o vou utilizar na an√°lise.**

```python
df.drop(['Review1', 'Review2', 'Review3'], axis=1, inplace=True)
print(df.head(10).to_string())
print(df.tail(10).to_string())
```

```python
df.describe()
```

**O primeiro gr√°fico √© sobre a coluna de Notas. Fiz primeiro um histograma para verificar poss√≠veis picos na distribui√ß√£o.**

```python
plt.figure(figsize=(8, 4))
plt.hist(df['Nota'], bins=30, color='red', alpha=0.8, edgecolor='pink')
plt.xticks(ticks=range(2, int(df['Nota'].max() + 1), 1))
plt.yticks(ticks=range(0, 250, 20))
plt.title('Histograma - Distribui√ß√£o de Notas')
plt.xlabel('Notas')
plt.ylabel('Frequ√™ncia')
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
plt.show()
```

**Em seguida, criei um gr√°fico de densidade a fim de melhor visualizar a distribui√ß√£o das notas.**

```python
plt.figure(figsize=(8, 4))
sns.kdeplot(df['Nota'], fill=True, color='red')
plt.title('Densidade de Notas')
plt.xlabel('Nota')
plt.ylabel('Densidade')
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
plt.show()
```

**Come√ßando a analisar a coluna de g√™nero, contei quantos valores √∫nicos ela tinha e pedi para listar quais eram eles.**

```python
print(f"N√∫mero de valores √∫nicos em 'G√™nero': {df['G√™nero'].nunique()}")
print(f"Lista de valores em 'G√™nero': {df['G√™nero'].unique()}")
```

**Notando que v√°rias dos valores eram sobrepostos, considerei ser melhor renomear alguns deles para diminuir o n√∫mero de categorias para o gr√°fico.**

```python
df['G√™nero'] = df['G√™nero'].replace('Unissex', 'Sem g√™nero')
df['G√™nero'] = df['G√™nero'].replace('Meninas', 'Feminino Infantil')
df['G√™nero'] = df['G√™nero'].replace('Meninos', 'Masculino Infantil')
df['G√™nero'] = df['G√™nero'].replace('Unissex', 'Sem g√™nero')
df['G√™nero'] = df['G√™nero'].replace('short menina verao look mulher', 'Feminino')
df['G√™nero'] = df['G√™nero'].replace('Mulher', 'Feminino')
df['G√™nero'] = df['G√™nero'].replace('roupa para gordinha pluss P ao 52', 'Feminino')
df['G√™nero'] = df['G√™nero'].replace('bermuda feminina brilho Blogueira', 'Feminino')
df['G√™nero'] = df['G√™nero'].replace('menino', 'Masculino Infantil')
```

**Agora, escolhi o gr√°fico de pizza para visualizar a propor√ß√£o dessas categorias.**

```python
plt.figure(figsize=(8, 4))
x = df['G√™nero'].value_counts().index
y = df['G√™nero'].value_counts().values
plt.pie(y, labels=x, autopct='%1.1f%%')
plt.title('Distribui√ß√£o de G√™nero')
plt.show()
```

**E um gr√°fico de barras para verificar a quantidade de cada uma das categorias.**

**Para a coluna de Marcas, primeiro verifiquei a quantidade de valores √∫nicos e os listei.**

```python
print(f"N√∫mero de valores √∫nicos em 'Marca': {df['Marca'].nunique()}")
print(f"Lista de valores em 'Marca': {df['Marca'].unique()}")
```

**Entretanto, o n√∫mero foi muito elevado. Assim, solicitei a uma IA para criar um script para tentar fazer uma redu√ß√£o de valores redundantes como feito anteriormente.**

```python
df['Marca'] = df['Marca'].replace(['roupa', 'vestu√°rio', 'roupas', 'conjunto', 'kit'], 'Roupa', regex=True)
df['Marca'] = df['Marca'].replace(['infantil', 'kids', 'baby'], 'Infantil', regex=True)
df['Marca'] = df['Marca'].replace(['fitness', 'academia', 'esportes', 'sport', 'sports'], 'Esporte', regex=True)
df['Marca'].replace({'multimarcas': 'V√°rias Marcas', 'variados': 'V√°rias Marcas'}, inplace=True)

# Corrigindo erros de digita√ß√£o (ajuste conforme necess√°rio)
df['Marca'] = df['Marca'].replace('puma', 'Puma', regex=True)

# Simplificando descri√ß√µes
df['Marca'] = df['Marca'].replace('camiseta dry fit masculinas', 'Camiseta', regex=True)
df['Marca'] = df['Marca'].replace('short feminino', 'Short', regex=True)

# Outros exemplos (adapte conforme sua necessidade):
df['Marca'] = df['Marca'].replace('jeans', 'Jeans', regex=True)
df['Marca'] = df['Marca'].replace('lingerie', 'Lingerie', regex=True)

# Substituindo termos muito gen√©ricos por "Outro"
df['Marca'] = df['Marca'].replace(['gen√©rica', 'pr√≥pria', 'nacional'], 'Outro', regex=True)

# Criando uma fun√ß√£o para substitui√ß√µes mais complexas
def custom_replace(text):
    if 'atacado' in text:
        return 'Atacado'
    elif 'baby' in text:
        return 'Infantil'
    else:
        return text

df['Marca'] = df['Marca'].apply(custom_replace)
```

```python
print(f"N√∫mero de valores √∫nicos em 'Marca': {df['Marca'].nunique()}")
```

**Como ainda assim o conjunto continuou grande, considerei melhor analisar somente as 10 mais vendidas.**
